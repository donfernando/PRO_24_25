````csharp
class Ejercicio
{
    public static bool EstaOrdenado(int[] array)
    {
        int i = 0;
        bool esDescendente = true;
        if (array !=null && array.Length > 1)
        {
            while (i < array.Length - 2 & array[i] >= array[i + 1])
                i++;
            esDescendente = array[i] >= array[i + 1];
        }
        return esDescendente;
    }    
}
````



### Análisis de Caja Blanca

1. **Entradas y Salidas**:
   - Entrada: Un array de enteros `array`.
   - Salida: Un valor booleano que indica si el array está ordenado en orden descendente según la lógica implementada.

2. **Caminos del Programa**:
   - **Caso 1**: El array tiene una longitud de 1 o menos.
     - En este caso, el programa no entra en el bloque `if` y devuelve `true`.
   - **Caso 2**: El array tiene más de un elemento.
     - El programa utiliza un bucle `while` que se ejecuta mientras `i` sea menor que `array.Length - 2` y `array[i] >= array[i + 1]`.
     - Después del bucle, se establece `esDescendente` basándose en si el último par de elementos comparados (`array[i]` y `array[i + 1]`) está en orden descendente.

3. **Puntos Críticos**:
   - La condición `array.Length > 1`, que determina si se entra en el bloque de código que verifica el orden.
   - El uso del operador `&` en lugar del operador `&&` de corte en la condición del `while`. Esto pone en riesgo la posibilidad de que al evaluar la expresión completa derive en una excepción por exceder los índices validos del array.
   - La variable `esDescendente` se establece basándose solo en el último par de elementos comparados, tras salir del bucle.

   - Entrada: `array = [5, 5, 5, 5]`.

   - Entrada: `array = [5, 3, 2, 1, 6]`.
   - Salida esperada: `false`.

   - Entrada: `array = [5, 4]`.
   - Salida esperada: `true`.
